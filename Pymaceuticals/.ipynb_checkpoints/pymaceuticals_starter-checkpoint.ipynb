```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data = pd.read_csv(clinical_trial_data_to_load)
# Combine the data into a single dataset
combined_data_file = pd.merge(clinical_trial_data,mouse_drug_data, how = 'left', on = 'Mouse ID')
# Display the data table for preview
combined_data_file.head(20)
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
groupby_drug_timepoint = combined_data_file.groupby(["Drug","Timepoint"])
mean_tumor_volume = groupby_drug_timepoint["Tumor Volume (mm3)"].mean().reset_index()
# Convert to DataFrame
mean_tumor_volume_df = pd.DataFrame(mean_tumor_volume)
# Preview DataFrame
mean_tumor_volume_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
standard_error_tumor = groupby_drug_timepoint["Tumor Volume (mm3)"].sem().reset_index()
# Convert to DataFrame
standard_error_tumor_volume_df = pd.DataFrame(standard_error_tumor)
# Preview DataFrame
standard_error_tumor_volume_df.head()
```

```python
#Error values for ploting errorbar
tum_error_cap= standard_error_tumor_volume_df.loc[(standard_error_tumor_volume_df["Drug"] == "Capomulin"), "Tumor Volume (mm3)"]
tum_error_inf= standard_error_tumor_volume_df.loc[(standard_error_tumor_volume_df["Drug"] == "Infubinol"), "Tumor Volume (mm3)"]
tum_error_ket= standard_error_tumor_volume_df.loc[(standard_error_tumor_volume_df["Drug"] == "Ketapril"), "Tumor Volume (mm3)"]
tum_error_pla= standard_error_tumor_volume_df.loc[(standard_error_tumor_volume_df["Drug"] == "Placebo"), "Tumor Volume (mm3)"]
tum_error_cap
```

```python
# Minor Data Munging to Re-Format the Data Frames

#Mean
tum_data_munging_mean = mean_tumor_volume_df.pivot(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")

# Preview that Reformatting worked
tum_data_munging_mean.head()
```

```python
# Generate the Plot (with Error Bars)

time = list(tum_data_munging_mean.index)

plt.errorbar(time, tum_data_munging_mean['Capomulin'],
             yerr=tum_error_cap, linestyle = "dotted", linewidth= 1 , label= "Capomulin", color = "green", marker = "X") 
plt.errorbar(time, tum_data_munging_mean['Infubinol'],
             yerr=tum_error_inf,linestyle = "dotted", linewidth= 1 , label= "Infubinol", color = "blue", marker = "P") 
plt.errorbar(time, tum_data_munging_mean['Ketapril'],
             yerr=tum_error_ket,linestyle = "dotted", linewidth= 1 , label= "Ketapril", color = "black", marker = "d")
plt.errorbar(time, tum_data_munging_mean['Placebo'],
             yerr=tum_error_pla,linestyle = "dotted", linewidth= 1 , label= "Placebo", color = "red", marker = "D")


plt.title("Tumor Response to Treatment")
plt.ylabel("Rumor Volume (mm3)")
plt.xlabel("Time (Days)")
plt.grid()
plt.legend(loc="best")

# Save the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_rates_count = groupby_drug_timepoint["Mouse ID"].count().reset_index()
survival_rates_count = survival_rates_count.rename(columns = {"Mouse ID": "Mouse Count"})

# Convert to DataFrame
survival_rates_count_df = pd.DataFrame(survival_rates_count)
# Preview DataFrame
survival_rates_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
#tum_data_munging_mean = mean_tumor_volume_df.pivot(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")
surv_rat_data_munging_count = survival_rates_count_df.pivot(index = "Timepoint", columns  = "Drug", values = "Mouse Count")
# Preview the Data Frame
surv_rat_data_munging_count.head()
```

```python
#Converting to percentage
rat_cap_per = surv_rat_data_munging_count['Capomulin']/surv_rat_data_munging_count['Capomulin'][0]*100
rat_inf_per = surv_rat_data_munging_count['Infubinol']/surv_rat_data_munging_count['Infubinol'][0]*100
rat_ket_per = surv_rat_data_munging_count['Ketapril']/surv_rat_data_munging_count['Ketapril'][0]*100
rat_pla_per = surv_rat_data_munging_count['Placebo']/surv_rat_data_munging_count['Placebo'][0]*100

rat_per_df = pd.DataFrame({
    'Capomulin':rat_cap_per,
    'Infubinol':rat_inf_per,
    'Ketapril':rat_ket_per,
    'Placebo':rat_pla_per
})
```

```python
# Generate the Plot (Accounting for percentages)
plt.plot(time,rat_per_df['Capomulin'], 
         linestyle = "dotted", linewidth= 1 , label= "Capomulin", color = "green", marker = "X")
plt.plot(time,rat_per_df['Infubinol'], 
         linestyle = "dotted", linewidth= 1 , label= "Infubinol", color = "blue", marker = "P")
plt.plot(time,rat_per_df['Ketapril'], 
         linestyle = "dotted", linewidth= 1 , label= "Ketapril", color = "black", marker = "d")
plt.plot(time,rat_per_df['Placebo'], 
         linestyle = "dotted", linewidth= 1 , label= "Placebo", color = "red", marker = "D")

plt.title("Survival During Treatment")
plt.ylabel("Survival Rate (%)")
plt.xlabel("Time (Days)")
plt.grid()
plt.legend(loc="best")

# Save the Figure

# Show the Figure
plt.show()
```
